# Задание 2

Рыцарь и дракон

Цель: создать небольшую игру, в которой вам необходимо играть за рыцаря и сразиться с драконом

Требования:

- Создание персонажа (имя, информация о доспехах, оружии, урон, здоровье)
- Управление персонажем и мини сюжет
- Создание дракона (Имя, информация о здоровье и уроне)
- Боевая система (нанесение и получение урона, урон должен быть случайным в заданном диапазоне)
- Реализовать бой между драконом и рыцарем

````python
from random import randint


def knight():
    knight_info = {'name': 'Святослав', 'armor': 100, 'weapon': 'Православный лук', 'damage': 50, 'health': 100}
    return knight_info


def drago():
    drago_info = {'name': 'Смауг', 'health': 500, 'damage': 60}
    return drago_info


def main_fight():
    knight1 = knight()
    drago1 = drago()

    print('Гордый рыцарь Святослав вооружившись православным луком решил отбить одинокую гору у ужасного дракона Смауга. Он подошёл к горе и задул в рог. На этот шум дракон вылетел из пещеры и начался бой:')

    while True:
        fight = input(
'''
Выберите действие:
1. Атаковать
2. Использовать зелье здоровья
3. Обновить броню
''')

        if fight == '1':
            drago1['health'] -= (knight1['damage'] + randint(-15,15))

            print('Дракон бьёт в ответ\n')

            if drago1['damage'] > knight1['armor'] :
                real_damage = (drago1['damage'] + randint(-15,15)) - knight1['armor']
                knight1['health'] -= real_damage
                knight1['armor'] = 0
            else:
                knight1['armor'] -= (drago1['damage'] + randint(-15,15))

            if drago1['health'] < 0:
                drago1['health'] = 0
            if knight1['health'] < 0:
                knight1['health'] = 0

            print('Броня Святослава:', knight1['armor'])
            print('Здоровье Святослава:', knight1['health'])
            print('Здоровье Дракона:', drago1['health'])

        elif fight == '2':
            knight1['health'] += 100
            print('Здоровье Святослава:', knight1['health'])
        elif fight == '3':
            knight1['armor'] = 100
            print('Броня Святослава:', knight1['armor'])

        if drago1['health'] <= 0:
            print('\nСвятослав одолел Смауга и забрал все его богатства')
            break
        elif knight1['health'] <= 0:
            print('\nСвятослав пал в бою с Драконом')
            break


main_fight()

````

Цель - создать менеджера команды Pokémon, который позволит пользователям:

- Добавлять покемонов в свою команду. (если такого покемона еще нет в команде)

Для данной задачи используйте: https://pokeapi.co/


````python
import requests
from random import randint


def send_request(url):
    response = requests.get(url)
    status_code = response.status_code
    if status_code == 200:
        return response.json()
    return None


def app_del_pokemon():
    quest = input('''
Нажмите 1, если хотите добавить покемона в команду
Нажмите 2. если хотите удалить покемона из команды
''')

    if quest == '1':
        name1 = input('Введите имя покемона: ').lower()
        if name1 not in team:
            team.append(name1)
        else:
            print('Данный покемон уже состоит в вашей команде')

    elif quest == '2':
        name2 = input('Введите имя покемона: ').lower()
        if name2 in team:
            team.remove(name2)
        else:
            print('Данного покемона нет в вашей команде')

    print('Список вашей команды:', *team)
    return team


def all_pokemons_info():
    for name in team:
        url = f'https://pokeapi.co/api/v2/pokemon/{name}/'
        data = send_request(url)

        print('Имя покемона:', data['name'])
        types = data['types']

        for tipe in types:
            for k,v in tipe.items():
                if k == 'type':
                    print('Тип покемона:', v['name'])

        print('Вес покемона:',data['weight'])
        print('Рост покемона:',data['height'])

        abilities = data['abilities']
        all_abilities = []

        for ability in abilities:
            for k,v in ability.items():
                if k == 'ability':
                    all_abilities.append(v['name'])

        print('Способности покемона:', all_abilities)


def one_pokemons_info():
    name = input('Введите имя нужного вам покемона: ').lower()
    check = 0

    for i in team:
        if name == i:
            check +=1
            url = f'https://pokeapi.co/api/v2/pokemon/{name}/'
            data = send_request(url)

            print('Имя покемона:', data['name'])

            types = data['types']
            for tipe in types:
                for k,v in tipe.items():
                    if k == 'type':
                        print('Тип покемона:', v['name'])

            print('Вес покемона:',data['weight'])
            print('Рост покемона:',data['height'])

            abilities = data['abilities']
            all_abilities = []

            for ability in abilities:
                for k,v in ability.items():
                    if k == 'ability':
                        all_abilities.append(v['name'])

            print('Способности покемона:', all_abilities)

    if check == 0:
        print('Такого покемона нет в команде')


def training_fight():#Я абсолютно не разбираюсь в покемонах, поэтмоу не представляю как проходят их бои. Поэтому путь всё решает рандом
    first_pokemon = input('Введите имя первого покемона: ').lower()

    if first_pokemon in team:
        second_pokemon = input('Введите имя второго покемона: ').lower()

        if second_pokemon in team:
            win = randint(1,2)
            if win == 1:
                print(f'Победил {first_pokemon}')
            elif win == 2:
                print(f'Победил {second_pokemon}')
        else:
            print('Такого покемона нет в вашей команде')

    else:
        print('Такого покемона нет в вашей команде')


team = []

while True:
    choice = input('''
Что вы хотети?
1. Добавить или удалить покемона
2. Вывести информацию о всех покемонах
3. Вывести информацию об 1 покемоне
4. Провести тренировочный бой между 2 покемонами
5. Закончить работу
''')
    if choice == '1':
        app_del_pokemon()
    elif choice == '2':
        all_pokemons_info()
    elif choice == '3':
        one_pokemons_info()
    elif choice == '4':
        training_fight()
    elif choice == '5':
        break
````
